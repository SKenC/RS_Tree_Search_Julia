```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15), dpi=50)
```

```python
#st = pd.read_csv("./exp_data/RS_15d_10000-1.csv")[:-1]
st = pd.read_csv("./exp_data/RS_15d_10000-1ALL2.csv")[:-1]
#non_st = pd.read_csv('hiteijo_accuracy_1K_100K.csv')[99900:-1]
st = st.drop("step", axis=1)
```

```python
from matplotlib.legend_handler import HandlerLine2D
names = st.columns.values

fig = plt.figure(figsize=(9,7))

ax = fig.add_subplot(1,1,1)

plt.xlabel("step", fontsize=15)
plt.ylabel("accuracy", fontsize=15)

marks = ['|', 'o', 's', 'x', 'D', ' ']
#print(st[names[1]][::10])
for i in range(len(names)):
    line, = plt.plot(st[names[i]][::20], marker=marks[i], mfc='white', markersize=7, markevery = 20, label=names[i])
    plt.legend(handler_map={line: HandlerLine2D(numpoints=1)}, fontsize=15)
    
# plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),
#           ncol=3, fontsize=13)
plt.show()
```
